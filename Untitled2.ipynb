```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://d17h27t6h515a5.cloudfront.net/topher/2017/October/59dd1c4c_tmdb-movies/tmdb-movies.csv")


# Display basic information about the dataset
print(df.info())

# Handling missing values
df = df.dropna()  # Drop rows with missing values

# Convert data types if needed
df['release_date'] = pd.to_datetime(df['release_date'])

# Extract relevant information
df['year'] = df['release_date'].dt.year

# Display the cleaned dataset
print(df.head())

```

```python
print(df.describe())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Distribution of Budget
sns.histplot(df['budget'], bins=20, kde=True)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

```

```python
# Example: Extracting the length of the movie title
df['title_length'] = df['original_title'].apply(len)

```

```python
# Example: Extracting the month from the release date
df['release_month'] = df['release_date'].dt.month
```

```python
# Example: Handling outliers in the budget column
Q1 = df['budget'].quantile(0.25)
Q3 = df['budget'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['budget'] >= Q1 - 1.5 * IQR) & (df['budget'] <= Q3 + 1.5 * IQR)]

```

```python
mean_budget = df['budget'].mean()
median_budget = df['budget'].median()
mode_budget = df['budget'].mode()[0]
```

```python
budget_range = df['budget'].max() - df['budget'].min()
budget_variance = df['budget'].var()
budget_std_dev = df['budget'].std()
```

```python
budget_skewness = df['budget'].skew()
budget_kurtosis = df['budget'].kurt()
```
